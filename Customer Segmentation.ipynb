```python
df['Last_Name'].fillna("Unknown", inplace=True)
```

```python
df['job_industry_category'].fillna("Not Specified", inplace=True)
```

```python
df['DOB'] = pd.to_datetime(df['DOB'], errors='coerce')
df['Calculated_Age'] = 2025 - df['DOB'].dt.year
```

```python
df.head()
```

```python
df['gender'] = df['gender'].str.strip().str.capitalize()
df['Owns_Car'] = df['Owns_Car'].map({'Yes': 1, 'No': 0})
```

```python
df = df[df['deceased_indicator'] != 'Y']
```

```python
df = df[df['Country'] == 'Australia']
```

```python
df['Mass_Affluent'] = df['wealth_segment'].isin(['Affluent Customer', 'High Net Worth'])

df['Age_Group'] = pd.cut(df['AGE'], bins=[0, 29, 45, 60, 100],
                         labels=['<30', '30-45', '46-60', '60+'])

df['Customer_Value'] = df['property_valuation'] * 10000 + df['past_3_years_bike_related_purchases'] * 200
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Wealth segment count
sns.countplot(data=df, x='wealth_segment')
plt.title("Wealth Segment Distribution")

# Age vs. Value
sns.scatterplot(data=df, x='AGE', y='Customer_Value', hue='Mass_Affluent')
```

```python
df['Is_Mass_Affluent_Custom'] = (
    (df['property_valuation'] >= 10) |  # Properties worth $100,000+
    (df['past_3_years_bike_related_purchases'] >= 10)
) & (df['AGE'] >= 30) & (df['AGE'] <= 65)
```

```python
features = df[['AGE', 'tenure', 'past_3_years_bike_related_purchases', 'property_valuation', 'Owns_Car']]
```

```python
features['Owns_Car'] = df['Owns_Car'].map({'Yes': 1, 'No': 0})
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(df[['AGE', 'property_valuation', 'past_3_years_bike_related_purchases']], diag_kind='kde')
```

```python
# One-hot encode key categorical columns
categorical_cols = ['State', 'job_industry_category', 'wealth_segment']
df_encoded = pd.get_dummies(df, columns=categorical_cols, drop_first=True)
```

```python
features = [
    'AGE', 'tenure', 'past_3_years_bike_related_purchases', 'property_valuation', 'Owns_Car'
] + [col for col in df_encoded.columns if col.startswith('State_') or col.startswith('job_') or col.startswith('wealth_')]

X = df_encoded[features]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=0.95)  # Keep 95% variance
X_pca = pca.fit_transform(X_scaled)

print(f"Reduced to {X_pca.shape[1]} components")
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inertia = []
cluster_range = range(2, 11)

for k in cluster_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)  # or X_pca if using PCA
    inertia.append(kmeans.inertia_)

# Plot the elbow graph
plt.plot(cluster_range, inertia, marker='o')
plt.xlabel("Number of Clusters")
plt.ylabel("Inertia (Within-Cluster SSE)")
plt.title("Elbow Method for Optimal k")
plt.show()
```

```python
kmeans = KMeans(n_clusters=4, random_state=42)
df['Cluster'] = kmeans.fit_predict(X_scaled)  # or X_pca
```

```python
# Group and compare mean stats
df.groupby('Cluster')[['AGE', 'tenure', 'property_valuation',
                       'past_3_years_bike_related_purchases']].mean().round(1)
```
